```python
idCols = ['passer_id','rusher_id','receiver_id']
valCols = [col for col in df.columns if col not in idCols]
```

```python
df.set_index(valCols).stack().reset_index().drop('play_id',axis=1).groupby(['season','game_id',0]).sum()
```
