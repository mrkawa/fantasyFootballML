```python
gamePass = pd.read_csv('../data/nflscrapRDownload/game_passing_df.csv').rename(columns={'Passer_ID':'Player_ID'})
gameRec = pd.read_csv('../data/nflscrapRDownload/game_receiving_df.csv').rename(columns={'Receiver_ID':'Player_ID'})
gameRush = pd.read_csv('../data/nflscrapRDownload/game_rushing_df.csv').rename(columns={'Rusher_ID':'Player_ID'})
seasonPass = pd.read_csv('../data/nflscrapRDownload/season_passing_df.csv').rename(columns={'Passer_ID':'Player_ID'})
seasonRec = pd.read_csv('../data/nflscrapRDownload/season_receiving_df.csv').rename(columns={'Receiver_ID':'Player_ID'})
seasonRush = pd.read_csv('../data/nflscrapRDownload/season_rushing_df.csv').rename(columns={'Rusher_ID':'Player_ID'})
```

```python
game = gameRec.merge(gameRush, how='outer', on=['GameID','Player_ID','Player_Name','Team','Opponent'], suffixes=['_rec','_rush']).merge(gamePass, how='outer', on=['GameID','Player_ID','Player_Name','Team','Opponent'], suffixes=['_rush','_pass'])
game = game[[col for col in game.columns if not any(x in col for x in ['WPA','EPA','wpa','epa','Success_Rate'])]]

season = seasonRec.merge(seasonRush, how='outer', on=['Season','Player_ID','Player_Name'], suffixes=['_rec','_rush']).merge(seasonPass, how='outer', on=['Season','Player_ID','Player_Name'], suffixes=['_rush','_pass'])
season = season[[col for col in season.columns if not any(x in col for x in ['WPA','EPA','wpa','epa','Success_Rate'])]]
```
